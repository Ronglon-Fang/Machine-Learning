### Example: classification and regression

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xp=np.array([2,3,3.5,4.5,4.7,5, 8])
yp=np.ones(np.shape(xp)[0])
xm=np.array([-1,-2,-3.5,-4,-4.5,-8])
ym=-np.ones(np.shape(xm)[0])
vx=np.concatenate((xm,xp))
vy=np.concatenate((ym,yp))
```

```python
fig,ax=plt.subplots()
ax.plot(xp,yp,color='orange', marker='+', linestyle='none', linewidth=2, markersize=12)
ax.plot(xm,ym,color='blue', marker='x', linestyle='none', linewidth=2, markersize=12)
#ax.grid(True)

# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')
# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
# set the y-spine
ax.spines['bottom'].set_position('zero')
# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

ax.axis([-9,9,-2,2])
ax.set_aspect('equal')

plt.savefig('lec8_1.eps', bbox_inches='tight')

plt.show(ax)
```

```python
N=vx.shape[0]
mA=np.array([vx,np.ones(N)])
mA=np.transpose(mA)
```

```python
vw=LA.solve(np.transpose(mA)@mA, np.transpose(mA)@vy)
print(vw)
```

```python
def f(vw, x):
    return vw[0]*x + vw[1] 
```

```python
xx=np.linspace(-8,8,200)
yy=f(vw,xx)

fig,ax=plt.subplots()
ax.plot(xp,yp,color='orange', marker='.', linestyle='none', linewidth=2, markersize=12)
ax.plot(xp,np.zeros(xp.shape[0]),color='orange', marker='+', linestyle='none', linewidth=2, markersize=12)
ax.plot(xm,ym,color='blue', marker='.', linestyle='none', linewidth=2, markersize=12)
ax.plot(xm,np.zeros(xm.shape[0]),color='blue', marker='*', linestyle='none', linewidth=2, markersize=12)
ax.plot(xx,yy,color='black',linewidth=2)
ax.plot([6,-3],[0,0],color='red', marker='x', linestyle='none', linewidth=2, markersize=12)

# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')
# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
# set the y-spine
ax.spines['bottom'].set_position('zero')
# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

ax.axis([-9,9,-2.5,2.5])
ax.set_aspect('equal')
ax.annotate('$f(x)=0.2x-0.04$', xy=(6, 1.2), xytext=(6.5, 1.8))

plt.savefig('lec8_2.eps',bbox_inches='tight')

plt.show()


fig,ax=plt.subplots()
ax.plot(xp,np.zeros(xp.shape[0]),color='orange', marker='+', linestyle='none', linewidth=2, markersize=12)
ax.plot(xm,np.zeros(xm.shape[0]),color='blue', marker='*', linestyle='none', linewidth=2, markersize=12)
ax.plot(0.2*np.ones(200),np.linspace(-1,1,200),color='black',linestyle='dashed',linewidth=2)


# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')
# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
# set the y-spine
ax.spines['bottom'].set_position('zero')
# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

ax.axis([-9,9,-2.5,2.5])
ax.set_aspect('equal')
ax.annotate('$x-0.2=0$', xy=(6, 1.2), xytext=(0.5, 0.8))
ax.annotate('$x-0.2>0$', xy=(6, 1.7), xytext=(1.5, 1.5), arrowprops = dict(facecolor='orange', arrowstyle="->"))
ax.annotate('$x-0.2<0$', xy=(-6, 1.7), xytext=(-3, 1.5), arrowprops = dict(facecolor='blue', arrowstyle="->"))
ax.set_title('1D separation')


plt.savefig('lec8_3.eps',bbox_inches='tight')

plt.show()
```
